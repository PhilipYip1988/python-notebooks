```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x1 = np.array([1, 2])
y1 = np.array([3, 6])

fig = plt.figure()
ax2d = fig.add_subplot(1, 1, 1)
ax2d.plot(x1, y1, marker='o')
ax2d.set_xlabel('x1')
ax2d.set_ylabel('y1')

print(ax2d, type(ax2d))
```

```python
import matplotlib.pyplot as plt
import numpy as np

r = np.arange(0, 2, 0.01)
theta = 2 * np.pi * r

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, r)
ax.set_rmax(2)
ax.set_rticks([0.5, 1, 1.5, 2])  # Less radial ticks
ax.set_rlabel_position(-22.5)  # Move radial labels away from plotted line
ax.grid(True)

ax.set_title("A line plot on a polar axis", va='bottom')
plt.show()
```

```python
r = np.arange(0, 1, 0.1)
theta = 2 * np.pi * r

degree = np.rad2deg(theta)
y = r * np.sin(theta)
x = r * np.cos(theta)

pd.DataFrame({'x': x, 'y': y, 'r': r, 'theta': theta, 'degree': degree, })
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, marker='o')

fig = plt.figure()
axpolar = fig.add_subplot(1, 1, 1, projection='polar')
axpolar.plot(theta, r, marker='o')
```
