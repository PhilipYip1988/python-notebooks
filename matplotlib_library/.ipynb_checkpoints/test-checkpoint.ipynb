```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(facecolor='lightyellow', edgecolor='black', linewidth=10);
ax = fig.add_subplot();
ax.set_visible(b=False);
fig.set_figwidth();
```

```python
fig, ax = plt.subplots(facecolor='lightyellow', edgecolor='black', linewidth=10);
fig.set_figwidth(4)
```

```python
fig = plt.figure(facecolor='lightyellow', edgecolor='black', linewidth=10);
ax = fig.add_axes([0, 0, 0.9, 0.9], facecolor='lightcyan');
ax.scatter(x=np.array([1, 2, 3, 4, 5]), 
           y=np.array([2, 4, 6, 8, 10]), color='r');
fig.set_figwidth(4);
```
