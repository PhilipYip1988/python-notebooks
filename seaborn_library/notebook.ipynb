## Style

The ```seaborn``` library has the function ```set_style``` which will change the style used for a ```matplotlib``` plot and by extension a ```seaborn``` plot. There are four styles ```'dark'```, ```'darkgrid```, ```'white'``` and ```'whitegrid'```. Normally the style is set after importing the libraries and made consistent for all the plots in a notebook:

````python
x = np.array([0, 1, 2, 3, 4])

sns.set_style('dark')
fig, ax = plt.subplots()
ax.plot(x, x, color='royalblue')

sns.set_style('darkgrid')
fig, ax = plt.subplots()
ax.plot(x, x, color='royalblue')

sns.set_style('white')
fig, ax = plt.subplots()
ax.plot(x, x, color='royalblue')

sns.set_style('whitegrid')
fig, ax = plt.subplots()
ax.plot(x, x, color='royalblue')
````

## Palette

Instead of manually selecting the color of each line individually, a color palette is typically used. ```sns``` has a ```palettes``` module which groups together color palette related identifiers:

````python
sns.palettes
````

The ```sns.palettes.color_palette``` function (also available as ```sns.color_palette```) will display a ```sns``` or ```mpl``` palette of discrete colors. The ```sns``` default color palette ```'pastel'``` uses a selection of Crayola crayon colors that are close to the colors in the ```mpl``` default palette ```'tab10'```:

````python
sns.color_palette('pastel')
````

````python
sns.color_palette('tab10')
````

There are also some slight variations of this colour palette:

````python
sns.color_palette('muted')
````

````python
sns.color_palette('dark')
````

````python
sns.color_palette('bright')
````

````python
sns.color_palette('colorblind')
````

Notice each of the discrete color palettes above have 10 colors. If the following for loop can be used to produce 15 lines:

````python
fig, ax = plt.subplots()
for num in range(15):
    ax.plot(x, x+num, lw=5)
````

Notice since the default color palette ```'tab10'``` only has 10 colors and there are 15 lines, the additional 5 lines reuse the colors from the beginning of the palette.

The color palette ```'tab20``` can be viewed:

````python
sns.color_palette('tab20')
````

The ```sns.palettes.set_palette``` function also available as (```sns.set_palette```) can be used to change the color palette to ```'tab20'``` and now each line will have a unique color:

````python
sns.set_palette(palette='tab20')
fig, ax = plt.subplots()
for num in range(15):
    ax.plot(x, x+num, lw=5)
````

A custom palette can also be made using the function ```sns.palettes.blend_palette``` (also available as ```sns.blend_palette```) and supplying the input argument ```colors``` which is usually a ```tuple``` of color strings, where each string is a CSS4 color natively supported by matplotlib and ```n_colors``` is an integer corresponding to the number of colors to be added to the palette:

````python
sns.blend_palette(colors=('royalblue', 'darkseagreen', 'yellow', 'tomato'), n_colors=15)
````

When only 2 values in ```colors``` are specified in the blend palette, the pattern can often be more sequential:

````python
sns.blend_palette(colors=('royalblue', 'tomato'), n_colors=15)
````

This can be set to the palette for a plot and notice how the color palette used complements the ```+1``` upward trend of each line:

````python
blend_palette15 = sns.blend_palette(colors=('royalblue', 'tomato'), n_colors=15)
sns.set_palette(palette=blend_palette15)
fig, ax = plt.subplots()
for num in range(15):
    ax.plot(x, x+num, lw=5)
````

The ```sns.palettes.xkcd_palette``` (also available as ```sns.xkcd_palette```) can be used to create a color palette from these colors. This function requires every color in the palette to be specified as a recognised key in the ```dict``` instance ```xkcd_rgb```:

## DataFrame Data Structure

```mpl``` is generally configured to take data in, in the form of a ```np.ndarray``` (1D):

````python
x = np.array([0, 1, 2, 3, 4])
y = np.array([0, 2, 4, 6, 8])
plt.plot(x, y);
````

A ```pd.Series``` can also be used, which recall is essentially a ```np.ndarray``` (1D) with a label:

````python
df = pd.DataFrame({'x': np.array([0, 1, 2, 3, 4]),
                   'y': np.array([0, 2, 4, 6, 8])})
plt.plot(df['x'], df['y']);
````

However no advantage is taken of the ```pd.Series``` name:

````python
df = pd.DataFrame({'x': np.array([0, 1, 2, 3, 4]),
                   'y': np.array([0, 2, 4, 6, 8])})
plt.plot(df['x'], df['y'])
plt.xlabel('x')
plt.ylabel('y');
````

```sns``` plotting functions on the other hand are generally configured with the input argument ```data``` which is provided as a ```pd.DataFrame``` and each axis is assigned to the respective ```pd.Series``` name in the for of a ```str```. These names will also automatically be used for the perspective axis label when not otherwise specified:

````python
df = pd.DataFrame({'x': np.array([0, 1, 2, 3, 4]),
                   'y': np.array([0, 2, 4, 6, 8])})
sns.lineplot(data=df, x='x', y='y');
````

If a categorical ```pd.Series``` is added:

````python
df = pd.DataFrame({'x': np.array([0, 1, 2, 3, 4]),
                   'y': np.array([0, 2, 4, 6, 8]),
                   'z': np.array(['even', 'odd', 'even', 'odd', 'even'])})
df['z'] = df['z'].astype('category')
````

The ```pd.DataFrame``` looks like the following:

````python
df
````

If the ```pd.DataFrame``` is split by the categorical ```pd.Series```, then notice there are two $x, y$ pairs:

````python
df[df['z'] == 'even']
````

````python
df[df['z'] == 'odd']
````

These can be plotted in ```mpl``` using the somewhat cumbersome syntax:

````python
plt.scatter(df['x'][df['z'] == 'even'], df['y'][df['z'] == 'even'])
plt.scatter(df['x'][df['z'] == 'odd'], df['y'][df['z'] == 'odd'])
plt.xlabel('x')
plt.xlabel('y');
````

In ```sns``` the syntax is much simpler and uses the ```hue``` parameter, notice also that a legend is also automatically generated in ```sns``` because multiple traces are present on the plot:

````python
sns.scatterplot(data=df, x='x', y='y', hue='z');
````

## AxesLevel and FigureLevel

The ```flights``` dataset is relational showing the number of flights with respect to time. If the ```sns.relational``` module is examined notice that there are three plotting functions ```lineplot```, ```scatterplot``` and ```relplot```:

````python
print_identifier_group(sns.relational, kind='all')
````

These plots are also available directly from ```sns```. Notice when ```sns.lineplot``` and ```sns.scatterplot``` are used that a ```mpl``` ```Axes``` instance is returned:

````python
sns.lineplot(data=flights, x='year', y='passengers')
````

````python
sns.scatterplot(data=flights, x='year', y='passengers')
````

The ```sns.relplot``` looks very similar to the ```sns.scatterplot``` at first glance:

````python
sns.relplot(data=flights, x='year', y='passengers')
````

Notice however that the plot above is one of the three ```sns``` classes the ```sns.FacetGrid```. The three ```sns``` classes operate on the ```Figure``` level:

````python
print_identifier_group(sns, kind='all')
````

The Figure level plotting function ```sns.relplot``` has a parameter ```kind``` which can be used to change the kind of plot to the ```Axes``` level equivalents. For ```sns.relplot``` the default value is ```'scatter'``` but can be changed to ```'line'```:

````python
sns.relplot(data=flights, x='year', y='passengers', kind='line')
````

A plotting function which outputs a ```sns.FacetGrid``` contains the additional parameters ```row``` and ```col``` which can be assigned to additional categorical ```pd.Series``` names, alongside ```col_wrap``` which can be used to set the number of plots for each column when the ```row``` parameter is not set. Now it is easy to see why a ```FacetGrid``` operates on the ```Figure``` level as the plot contains multiple ```Axes```:

````python
sns.relplot(data=flights, x='year', y='passengers', hue='month', col='month', col_wrap=3, kind='line')
````

The ```FacetGrid``` is ```Figure``` level and is designed to be used as a standalone ```Figure```. The ```Axes``` level plot can be used with other ```mpl``` plotting functions or ```sns``` ```Axes``` level plotting functions and has the ```ax``` parameter which can be used to select an ```ax``` to plot to:

````python
fig, ax = plt.subplots(2, 1, figsize=(12, 10))
sns.scatterplot(data=flights, x='year', y='passengers', hue='month', ax=ax[0])
sns.lineplot(data=flights, x='year', y='passengers', hue='month', ax=ax[1])
````

````python
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
sns.scatterplot(data=flights, x='year', y='passengers', hue='month', ax=ax)
sns.lineplot(data=flights, x='year', y='passengers', hue='month', ax=ax)
````

If only April flights are examined:

````python
april_flights = flights[flights['month'] == 'Apr']
````

````python
april_flights
````

If the ```sns.scatterplot``` for April flights is examined, notice that there is a visual trend that is approximately a linear relation and in many cases it is useful to add a regression line to view the relationship:

```sns.regression``` can be used for plots that contain a regression curve:

````python
print_identifier_group(sns.regression, kind='all')
````

Notice there are three main plotting functions ```sns.regplot``` and ```sns.residplot``` operate on the ```Axes``` level and ```sns.lmplot``` operates on the ```Figure``` level. If the ```Axes``` level plots are examined first, a 1st, 2nd and 3rd order polynomial fit can be fitted to the data and plotted alongside the residuals:

````python
fig, ax = plt.subplots(2, 1, figsize=(12, 8))
sns.regplot(data=april_flights, x='year', y='passengers', ax=ax[0]);
sns.residplot(data=april_flights, x='year', y='passengers', ax=ax[1]);
sns.regplot(data=april_flights, x='year', y='passengers', order=2, ax=ax[0]);
sns.residplot(data=april_flights, x='year', y='passengers', order=2, ax=ax[1]);
sns.regplot(data=april_flights, x='year', y='passengers', order=3, ax=ax[0]);
sns.residplot(data=april_flights, x='year', y='passengers', order=3, ax=ax[1]);
````

The linear regression line suffices for this dataset and has random residuals. A small change occurs moving over to a quadratic regression line and there is a very small difference between a quadratic and a cubic regression.

```sns.lmplot``` is the Figure level equivalent to ```sns.regplot``` however is more limited and locked to a linear model:

````python
sns.lmplot(data=april_flights, x='year', y='passengers')
````

The ```Figure``` level ```FacetGrid``` does have the advantage of being able to use ```col``` and ```row``` to create subplots for a categorical ```pd.Series```:

````python
sns.lmplot(data=flights, x='year', y='passengers', hue='month', col='month', col_wrap=3)
````

The ```sns.distribution``` module groups together distribution plots. The ```sns.histplot```, ```sns.rugplot```, ```sns.kdeplot``` and ```sns.ecdfplot``` are ```Axes``` level and ```sns.displot``` is ```Figure``` level outputting a ```FacetGrid``` respectively:

````python
print_identifier_group(sns.distributions, kind='all')
````

```sns.histplot``` will be used to create a histogram. To understand these plotting functions better only ```x``` will be assigned to a ```pd.Series``` with name ```'petal_length'```and ```hue``` will be assigned to a ```pd.Series``` with name ```'species'```, the ```bins``` will be automatically selected:

````python
sns.histplot(data=iris, x='petal_length', hue='species');
````

However a larger number of bins can be used, for example 100. Visually it is easy to distinguish the blue distribution setosa from the orange distribution vesicolor from the green distribution virginica. The orange distribution vesicolor and green distribution virginica can also be visually distinguished from one another but have an overlap region:

````python
sns.histplot(data=iris, x='petal_length', hue='species', bins=100);
````

```sns.rugplot``` will be used to create a rugplot. A line is essentially drawn for each datapoint at the bottom of the figure, this creates a rug for each distribution. Notice when many datapoints overlap, the lines in the rug appear to be darker:

````python
sns.rugplot(data=iris, x='petal_length', hue='species');
````

```sns.kdeplot``` will be used to create a kernel density estimation plot. For each datapoint instead of a line, a Gaussian is made. The Gaussians for each datapoint are then summed creating a kde:

````python
sns.kdeplot(data=iris, x='petal_length', hue='species');
````

The ```sns.ecdf``` plot creates an empirical cumulative distribution function. The ECDF plots out the proportion of datapoints at and below a value, starting at 0 and ending at 1. Again it can be used to visually distinguish the three distributions:

````python
sns.ecdfplot(data=iris, x='petal_length', hue='species');
````

```sns.histplot``` and ```sns.kdeplot``` also have the parameter ```y``` which allows the plot to be expanded to 2 dimensions, with the counts or density shown as a colormap:

````python
sns.histplot(data=iris, x='petal_length', y='sepal_width', hue='species', bins=50);
````

It is quite common to use the rug plot alongside the kde plot in 2D:

````python
sns.rugplot(data=iris, x='petal_length', y='petal_width', hue='species');
sns.kdeplot(data=iris, x='petal_length', y='petal_width', hue='species');
````

```sns.categorical``` contains a number of categorical plotting functions. ```sns.stripplot```, ```sns.swarmplot```, ```sns.boxplot```, ```sns.violinplot```, ```sns.pointplot```, ```sns.barplot``` are ```Axes``` level and ```catplot``` is ```Figure``` level outputting a ```FacetGrid```:

````python
print_identifier_group(sns.categorical, kind='all')
````

The ```sns.stripplot``` and ```sns.swarmplot``` are both categorical scatter plots, there are slight differences in the way the datapoints are laid out in order to make the trend more visible:

````python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
sns.stripplot(data=exercise, x='pulse', y='kind', hue='diet', ax=ax[0])
sns.swarmplot(data=exercise, x='pulse', y='kind', hue='diet', ax=ax[1])
````

The ```sns.boxplot``` visually summarises the physical attribute distribution using a mean value, enclosed in an inner quartile range box that has whiskers which span out to the minimum and maximum values (excluding outliers). The inner quartile range contains the closest 50 % of values to the mean (1 standard deviation). The violin plot is similar but makes a KDE instead of a box. Normally the KDE is mirrored around the top and bottom of the box however when there are only two categories it can be split as shown:

````python
fig, ax = plt.subplots(2, 2, figsize=(12, 10))
sns.boxplot(data=exercise, x='pulse', y='kind', hue='diet', ax=ax[0, 0])
sns.violinplot(data=exercise, x='pulse', y='kind', hue='diet', ax=ax[0, 1])
sns.swarmplot(data=exercise, x='pulse', y='kind', hue='diet', ax=ax[1, 0])
sns.violinplot(data=exercise, x='pulse', y='kind', hue='diet', ax=ax[1, 1], split=True)
````

The ```sns.barplot``` and ```sns.pointplot``` are other simple representations of the distribution. The bar plot is typically projected from a mean value to the origin. It also has an errorbar of 1 standard deviation. The point plot essentially just plots the mean value as a point and shows the standard deviation around it:

````python
fig, ax = plt.subplots(2, 1, figsize=(12, 6))
sns.barplot(data=exercise, x='pulse', y='kind', hue='diet', ax=ax[0])
sns.pointplot(data=exercise, x='pulse', y='kind', hue='diet', ax=ax[1])
````
