# The seaborn data visualisation library

Seaborn is a data visualisation library based upon ```matplotlib```. It has an application interface which simplifies the creation of complex but common visualisation plots from data contained within a ```pandas``` ```DataFrame``` instance. 

In theory all of the plots created by ```seaborn``` can be done directly in ```matplotlib``` but generally this requires much more lines of code.

More details about colormaps are given in Matplotlibs documentation [Matplotlib: Colormaps](https://matplotlib.org/stable/tutorials/colors/colormaps.html)

The following for loop can be used to produce 15 lines:

````python
fig, ax = plt.subplots()
for num in range(15):
    ax.plot(x, y+num, lw=5)
````

Since the color palette only has 10 colors and there are 15 lines, the additional 5 lines reuse the colors from the beginning of the palette.

The ```seaborn``` function ```set_palette``` can be used to set the default color palette to ```'pastel'```:

````python
sns.set_palette(palette='pastel')
````

````python
fig, ax = plt.subplots()
for num in range(15):
    ax.plot(x, y+num, lw=5)
````

A custom palette can also be made using the seaborn function ```blend_palette``` and supplying the input argument ```colors``` which is usually a ```tuple``` of color strings, where each string is a CSS4 color natively supported by matplotlib and ```n_colors``` is an integer corresponding to the number of colors to be added to the palette:

````python
bp = sns.blend_palette(colors=('tomato', 
                               'yellow', 
                               'darkseagreen', 
                               'royalblue'),
                       n_colors=10)
````

````python
bp
````

This custom palette can be set:

````python
sns.set_palette(palette=bp)
````

````python
fig, ax = plt.subplots()
for num in range(15):
    ax.plot(x, y+num, lw=5)
````

## Pandas DataFrame

The ```seaborn``` plotting functions generally assume the data is contained within an instance of the ```pandas``` class ```DataFrame```. 

The ```DataFrame``` instance for the simple plot types should generally have a numeric ```Series``` instance corresponding to $x$ and a numeric ```Series``` instance corresponding to $y$. For example:

````python
df = pd.DataFrame(data={'x': np.arange(0, 11, 1),
                        'y': np.arange(0, 22, 2)})
````

The ```DataFrame``` method ```head``` can be called from the instance ```df```:

````python
df.head()
````

The data type can be tested using the ```DataFrame``` attribute ```dtypes```:

````python
df.dtypes
````

If the ```DataFrame``` instance ```df``` has a categorical ```Series``` it can be used to plot subsets of the data using the category. A ```bool``` ```Series``` instance can be made, which in this case checks if the $x$ value is even:

````python
df['x'].head() % 2
````

This can be assigned to a new ```Series``` on the ```DataFrame``` instance ```df```:

````python
df['parity'] = df['x'] % 2 == 0
````

````python
df.head()
````

````python
df.dtypes
````

This ```Series``` instance can have its type changed to category and these categories can be renamed:

````python
df['parity'] = df['parity'].astype('category')
````

````python
df['parity'] = df['parity'].cat.rename_categories({True: 'even',
                                                   False: 'odd'})
````

The ```DataFrame``` instance ```df``` head can be checked:

````python
df.head()
````

````python
df.dtypes
````

The ```DataFrame``` instance ```df``` information can be checked:

````python
df.info()
````

````python
df.describe()
````

The ```seaborn``` plot functions are setup to expect data from a dataframe. The first few input arguments relate to the data and the other keyword arguments are optional, meaning if they are unspecified, the plot is carried out using the default values. The function ```print_identifier_group``` can be imported from the custom ```helper_module``` and used to view identifier groupings:

````python
from helper_module import print_identifier_group
````

The most common plotting functions can be accessed from the ```seaborn``` library directly for example the ```scatterplot```: 

````python
sns.scatterplot
````

However generally plotting functions are compartmentalised into their perspective modules. The ```scatterplot``` for example is a ```relational``` plot showing the relation between ```x``` and ```y``` values and is found in the ```relational``` module:

````python
print_identifier_group(sns.relational, kind='function')
````

There is also the ```regression``` module for regression plots:

```seaborn``` is a wrapper around ```pyplot``` and has plotting functions that operate on the ```Axes``` and ```Figure``` level respectively.

````python
type(sns.relational.scatterplot(data=df, x='x', y='y'))
````

Notice that the:

* the ```scatterplot``` is on the Axes level and is an instance of the ```Axes``` class from ```matplotlib```.

* ```relplot``` is the ```seaborn``` class ```FacetGrid``` which is essentially a modified ```Figure``` class from matplotlib. 

Recall that a ```Figure``` can contain multiple subplots.


````python
taxis.head()
````

The ```Figure``` level plotting function will by default use a single subplot and therefore the output will look similar to the Axes level plotting function:

````python
sns.relational.relplot(data=taxis, x='distance', y='fare')
````

````python
sns.relational.scatterplot(data=taxis, x='distance', y='fare')
````

However it can use a categorical series to split the data into seperate subplots:

````python
sns.relational.relplot(data=taxis, x='distance', y='fare', col='color')
````

````python
sns.relational.relplot(data=taxis, x='distance', y='fare', row='color')
````

The ```Axes``` level plots can be used with matplotlib plotting functions. Note however that the ```Axes``` class is from ```matplotlib``` and therefore has no access to ```seaborn``` plotting functions. Instead the ```ax``` parameter is supplied as an input argument and if not supplied the currently selected ```Axes``` is used:

````python
fig, ax = plt.subplots(3, 1)
ax[0].scatter(taxis['distance'], taxis['fare'])
sns.relational.scatterplot(data=taxis, x='distance', y='fare', ax=ax[1])
````

````python
fig, ax = plt.subplots(3, 1)
ax[0].scatter(taxis['distance'], taxis['fare'])
sns.relational.scatterplot(data=taxis, x='distance', y='fare')
````

Notice the difference when the ```Figure``` level ```relplot``` is used instead, it always returns a new ```Figure``` using the ```seaborn``` class ```AxisGrid```:

````python
fig, ax = plt.subplots(3, 1)
ax[0].scatter(taxis['distance'], taxis['fare'])

sns.relational.relplot(data=taxis, x='distance', y='fare')
````

If the following matplotlib plotting function and seaborn function are compared:

````python
fig, ax = plt.subplots(2, 1)
ax[0].scatter(taxis['distance'], taxis['fare'], s=30, lw=0.5, ec='white')
ax[0].set_xlabel('distance')
ax[0].set_ylabel('fare')

sns.relational.scatterplot(data=taxis, x='distance', y='fare', ax=ax[1])
````

Notice that the twosubplots look very similar however the code for the ```seaborn``` figure is simpler. Under the hood ```seaborn``` uses the ```scatter``` function of ```matplotlib```, applies styling and uses the name from each ```pd.Series``` for the labels.
