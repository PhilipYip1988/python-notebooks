# The seaborn data visualisation library

Seaborn is a data visualisation library based upon ```matplotlib```. It has an application interface which simplifies the creation of complex but common visualisation plots from data contained within a ```pandas``` ```DataFrame``` instance. 

In theory all of the plots created by ```seaborn``` can be done directly in ```matplotlib``` but generally this requires much more lines of code.

## Palette

The ```seaborn``` function ```color_palette``` can be used to view the color palette:

````python
sns.color_palette(palette='tab10')
````

````python
sns.color_palette(palette='pastel')
````

````python
sns.color_palette(palette='viridis')
````

More details about colormaps are given in Matplotlibs documentation [Matplotlib: Colormaps](https://matplotlib.org/stable/tutorials/colors/colormaps.html)

The following for loop can be used to produce 15 lines:

````python
fig, ax = plt.subplots()
for num in range(15):
    ax.plot(x, y+num, lw=5)
````

Since the color palette only has 10 colors and there are 15 lines, the additional 5 lines reuse the colors from the beginning of the palette.

The ```seaborn``` function ```set_palette``` can be used to set the default color palette to ```'pastel'```:

````python
sns.set_palette(palette='pastel')
````

````python
fig, ax = plt.subplots()
for num in range(15):
    ax.plot(x, y+num, lw=5)
````

A custom palette can also be made using the seaborn function ```blend_palette``` and supplying the input argument ```colors``` which is usually a ```tuple``` of color strings, where each string is a CSS4 color natively supported by matplotlib and ```n_colors``` is an integer corresponding to the number of colors to be added to the palette:

````python
bp = sns.blend_palette(colors=('tomato', 
                               'yellow', 
                               'darkseagreen', 
                               'royalblue'),
                       n_colors=10)
````

````python
bp
````

This custom palette can be set:

````python
sns.set_palette(palette=bp)
````

````python
fig, ax = plt.subplots()
for num in range(15):
    ax.plot(x, y+num, lw=5)
````

The ```seaborn``` library has the dictionary ```xkcd_rgb``` which has color name keys and hexadecimal values for the 954 xkcd colors:

````python
sns.xkcd_rgb
````

More details about these colors is available with the xkcd documentation [xkcd colors](https://xkcd.com/color/rgb/)

````python
fig, ax = plt.subplots()
ax.plot(x, y, color=sns.xkcd_rgb['sea green']);
````

The ```seaborn``` method ```xkcd_palette``` can be used to create a color palette from these colors. This function requires every color in the palette to be specified as one of the corresponding key names in the ```xkcd_rgb``` dictionary:

````python
xkcdp = sns.xkcd_palette(colors=('tomato red',
                                 'coral',
                                 'light plum',
                                 'cocoa',
                                 'yellowish', 
                                 'cool green',
                                 'faded green',
                                 'light teal',
                                 'faded blue',
                                 'cornflower blue'))
````

````python
xkcdp
````

The color palette can be set as the default:

````python
sns.set_palette(palette=xkcdp)
````

````python
fig, ax = plt.subplots()
for num in range(15):
    ax.plot(x, y+num, lw=5)
````

There are a number of functions such as ```choose_diverging_palette```, ```choose_light_palette```, ```choose_dark_palette```, ```choose_colorbrewer_palette```, ```choose_cube_helix_palette```, and ```choose_cubehelix_palette``` which use interactive Python widgets for a customisable color palette: 

````python
dp = sns.choose_diverging_palette()
````

The sliders should be adjusted to the desired settings before running the two cells below:

## Pandas DataFrame

The ```seaborn``` plotting functions generally assume the data is contained within an instance of the ```pandas``` class ```DataFrame```. 

The ```DataFrame``` instance for the simple plot types should generally have a numeric ```Series``` instance corresponding to $x$ and a numeric ```Series``` instance corresponding to $y$. For example:

````python
df = pd.DataFrame(data={'x': np.arange(0, 11, 1),
                        'y': np.arange(0, 22, 2)})
````

The ```DataFrame``` method ```head``` can be called from the instance ```df```:

````python
df.head()
````

The data type can be tested using the ```DataFrame``` attribute ```dtypes```:

````python
df.dtypes
````

If the ```DataFrame``` instance ```df``` has a categorical ```Series``` it can be used to plot subsets of the data using the category. A ```bool``` ```Series``` instance can be made, which in this case checks if the $x$ value is even:

````python
df['x'].head() % 2
````

This can be assigned to a new ```Series``` on the ```DataFrame``` instance ```df```:

````python
df['parity'] = df['x'] % 2 == 0
````

````python
df.head()
````

````python
df.dtypes
````

This ```Series``` instance can have its type changed to category and these categories can be renamed:

````python
df['parity'] = df['parity'].astype('category')
````

````python
df['parity'] = df['parity'].cat.rename_categories({True: 'even',
                                                   False: 'odd'})
````

The ```DataFrame``` instance ```df``` head can be checked:

````python
df.head()
````

````python
df.dtypes
````

The ```DataFrame``` instance ```df``` information can be checked:

````python
df.info()
````

The ```DataFrame``` instance ```df``` description can be checked which gives statistical details about only the numeric data:

````python
df.describe()
````

### Taxis Dataset

The taxis dataset contains information about passengers for three brands of taxis in Manhatten. This includes the pickup and dropoff times, the number of passengers, the distance, the fare, the tips, tolls paid, total cost, color (brand of taxi), payment method, pickup zone, dropoff zone, pickup borough and dropoff borough.

````python
taxis = sns.load_dataset(name='taxis')
````

````python
taxis.head()
````

````python
taxis.dtypes
````

````python
taxis.shape
````

````python
taxis.info()
````

Notice that some of the columns have non-null values. As this is a large dataset, it can be cleaned up by dropping the null values:

````python
taxis = taxis.dropna(how='any')
taxis = taxis.reset_index()
````

````python
taxis.info()
````

The ```Series``` that have the data type object can be converted to categorical:

````python
taxis['color'] = taxis['color'].astype('category')
taxis['payment'] = taxis['payment'].astype('category')
taxis['pickup_zone'] = taxis['pickup_zone'].astype('category')
taxis['dropoff_zone'] = taxis['dropoff_zone'].astype('category')
taxis['pickup_borough'] = taxis['pickup_borough'].astype('category')
taxis['dropoff_borough'] = taxis['dropoff_borough'].astype('category')
````

````python
taxis.info()
````

## Seaborn Plotting Functions

The ```seaborn``` plot functions are setup to expect data from a dataframe. The first few input arguments relate to the data and the other keyword arguments are optional, meaning if they are unspecified, the plot is carried out using the default values. The function ```print_identifier_group``` can be imported from the custom ```helper_module``` and used to view identifier groupings:

````python
from helper_module import print_identifier_group
````

The most common plotting functions can be accessed from the ```seaborn``` library directly for example the ```lineplot```: 

````python
sns.lineplot
````

However generally plotting functions are compartmentalised into their perspective modules. The ```lineplot``` for example is a ```relational``` plot showing the relation between ```x``` and ```y``` values and is found in the ```relational``` module:

````python
print_identifier_group(sns.relational, kind='method')
````

There is also the ```regression``` module for regression plots:

````python
sns.regression
````

````python
print_identifier_group(sns.regression, kind='method')
````

And a ```categorical``` module for categorical plots:


````python
sns.categorical
````
